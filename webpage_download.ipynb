```python
import pdfkit

def convert_webpage_to_pdf(url, output_path):
    # Define the path to wkhtmltopdf executable (installed in Colab environment)
    config = pdfkit.configuration(wkhtmltopdf='/usr/bin/wkhtmltopdf')

    try:
        # Convert the webpage to PDF
        pdfkit.from_url(url, output_path, configuration=config)
        print(f"Webpage saved to {output_path}")
    except Exception as e:
        print(f"Error converting webpage to PDF: {e}")
```

```python
# Example usage
url = 'https://learn.microsoft.com/en-us/azure/azure-functions/functions-app-settings'
output_path = '/content/output2.pdf'  # Save to a location in Colab's file system
convert_webpage_to_pdf(url, output_path)
```
